```python
# Sample tensors
a = torch.tensor([1, 2, 3, 4])
b = torch.tensor([2, 2, 1, 4])

print("a:", a)
print("b:", b)

print("\n--- Element-wise Comparisons ---")

print("a == b :", torch.eq(a, b))
print("a != b :", torch.ne(a, b))

print("a > b  :", torch.gt(a, b))
print("a >= b :", torch.ge(a, b))

print("a < b  :", torch.lt(a, b))
print("a <= b :", torch.le(a, b))

print("\n--- Using Operators (Same result) ---")

print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a >= b :", a >= b)
print("a < b  :", a < b)
print("a <= b :", a <= b)

print("\n--- Comparison with Scalars ---")

print("a > 2  :", a > 2)
print("a == 3 :", a == 3)

print("\n--- All / Any Conditions ---")

print("All a > 0 :", torch.all(a > 0))
print("All a > 0 :", torch.all(a > 2))
print("Any a > 3 :", torch.any(a > 3))
```

# *Inplace Operation*

```python
import torch

a = torch.tensor([1., 2., 3.])

a.add_(2)      # a = a + 2
print(a)

a.mul_(3)      # a = a * 3
print(a)
```

```python
a = torch.tensor([2., 4., 6.])

a.sub_(1)      # subtraction
a.div_(2)      # division
a.pow_(2)      # power

print(a)
```

```python
x = torch.tensor([-2., 0., 3.])

x.relu_()
print(x)
x.sigmoid_()
print(x)
x.tanh_()
print(x)
```

```python
t = torch.empty(3)

t.zero_()      # fill with zeros
t.fill_(5)     # fill with constant
print(t)
```

# *Copying and Cloning a Tensor*

```python
a = torch.tensor([1, 2, 3])
b = a          # NOT a copy
b[0] = 100
print(a)       # tensor([100, 2, 3])
```

```python
print(id(a))
print(id(b))
```

```python
a = torch.tensor([1., 2., 3.], requires_grad=True)

b = a.clone()

b[0] = 100
print(a)   # unchanged
print(b)
print(id(a))
print(id(b))
```

```python
import torch
import time

# Matrix size
N = 5_000

# ---------------- CPU ----------------
print("Running on CPU...")
A_cpu = torch.randn(N, N)
B_cpu = torch.randn(N, N)

start = time.time()
C_cpu = torch.matmul(A_cpu, B_cpu)
end = time.time()

print(f"CPU time: {end - start:.2f} seconds")

# ---------------- GPU ----------------
if torch.cuda.is_available():
    print("\nRunning on GPU...")
    device = torch.device("cuda")

    A_gpu = A_cpu.to(device)
    B_gpu = B_cpu.to(device)

    # warm-up (important for fair timing)
    torch.matmul(A_gpu, B_gpu)
    torch.cuda.synchronize()

    start = time.time()
    C_gpu = torch.matmul(A_gpu, B_gpu)
    torch.cuda.synchronize()   # wait for GPU to finish
    end = time.time()

    print(f"GPU time: {end - start:.2f} seconds")
else:
    print("\nCUDA not available")
```

# *Reshaping Tensor*

```python
import torch

# 2D tensor
x = torch.tensor([
    [1, 2, 3],
    [4, 5, 6]
])

print("Original tensor:")
print(x)
print("Shape:", x.shape)   # (2, 3)

# ---------------- reshape ----------------
r = x.reshape(3, 2)
print("\nReshape (2,3) -> (3,2):")
print(r)
print("Shape:", r.shape)

# ---------------- flatten ----------------
f = torch.flatten(x)
print("\nFlatten:")
print(f)
print("Shape:", f.shape)

# ---------------- permute ----------------
p = x.permute(1, 0)
print("\nPermute (transpose):")
print(p)
print("Shape:", p.shape)

# ---------------- unsqueeze ----------------
u = x.unsqueeze(0)
print("\nUnsqueeze at dim=0:")
print(u)
print("Shape:", u.shape)   # (1, 2, 3)

# ---------------- squeeze ----------------
s = u.squeeze(0)
print("\nSqueeze back:")
print(s)
print("Shape:", s.shape)   # (2, 3)
```

# pytorch and *numpy*

```python
# Memory is Shared. Change in One will be reflected to the other one
import numpy as np
import torch

np_arr = np.array([[1, 2], [3, 4]], dtype=np.float32)

t = torch.from_numpy(np_arr)
print(t)
```

```python
# Memory is Shared. Change in One will be reflected to the other one
t = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)

np_arr = t.numpy()
print(np_arr)
```
